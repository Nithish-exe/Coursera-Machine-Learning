```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle') # just the course theme for plots
```

```python
# x_train and y_train are the given/known points on a graph
# each value in the x_train array corresponds to the value in the y_train array
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train} \ny_train = {y_train}")
# the shape of a numpy array is a tuple with values for each dimension
# in a 1d array, the output is (# of elements,)
# in a 2d array, the output is (# of rowns, #number of colums)
print(f"x_train.shape = {x_train.shape}")
```

```python
# m is the number of training examples or the length of x_train and y_train
m = len(x_train)
print(f"Number of training examples: {m}")
```

```python
# (x^i, y^i) is the ith training example
# Python starts this list at zero so (x^0, y^0) is [1.0, 2.0]
i = 0 # change to 1 to see (x^1, y^1)
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{1}, y^{i}) = ({x_i}, {y_i})")
```

```python
# plot the graph for a visual
# parameters are x and y input. Marker and c are visual graph styles
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("Size in 1000s of square feet")
plt.show()
```

```python
# w is weightage and b is bias, different w and b values give you different lines on the graph
# you want to change the w and b values so they fit very close to your model
w = 200   # value after testing
b = 100   # value after testing
print(f"w = {w} \nb = {b}")
# you want to predict each point on the regression by using the formula x^i=fwb=w*x[i]+b
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):  
        f_wb[i] = w * x[i] + b # notice the formula is similar to the slope intercept formula
    return f_wb
# now we can store this output in a variable
tmp_f_wb = compute_model_output(x_train, w, b)
```

```python
# plot a new graph with the model's predictions
# fine tune w and b to make the prediction as close to the given data points as possible
plt.plot(x_train, tmp_f_wb, c='b', label='Model Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title("Housing Prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("Size in 1000s of square feet")
plt.legend()
plt.show()
```

```python
# now we can predict the cost of a house based on any input x
my_house_sqft = 1200   
cost_my_house = w * (my_house_sqft/1000) + b # remember sqft is in thousands
print(f"The cost of a house with {my_house_sqft} square feet is {cost_my_house} thousand dollars!")
```
